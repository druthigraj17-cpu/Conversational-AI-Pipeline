2.Local LLM Intialization and Research Query Testing.

```python
from transformers import pipeline

research_ai = pipeline(
    "text-generation",
    model="distilgpt2"
)

query = "Explain the importance of AI orchestration in conversational AI systems."

response = research_ai(
    query,
    max_new_tokens=100
)

print(response[0]['generated_text'])
```
