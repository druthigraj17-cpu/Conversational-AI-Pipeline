2.Loading the local model

```python
from transformers import pipeline

browser_agent = pipeline(
    "text-generation",
    model="distilgpt2"
)
```

3.Simulating the browser task (example)

```python
task = """
Open LinkedIn and search for AI engineering internships.
"""

response = browser_agent(
    f"Browser Agent Task: {task}\nStep-by-step browser actions:",
    max_new_tokens=120
)

print(response[0]['generated_text'])
```
